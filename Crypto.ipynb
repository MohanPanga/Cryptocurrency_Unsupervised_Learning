```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Resources/crypto_data.csv")
data
```

```python
data_trading =  data[data.IsTrading == True].copy(deep=True)
print(data_trading.shape)
print(data_trading.isnull().sum())
data_trading.dropna(inplace=True)
data_trading
```

### Data rows reduced to 685 after removing NaN rows

```python
data_trading.drop('IsTrading', axis=1, inplace=True)
data_trading
```

```python
data_trading_mined = data_trading[data_trading["TotalCoinsMined"] > 0].copy(deep=True)
data_trading_mined
```

```python
import seaborn as sns
data_plot = data_trading_mined.sort_values(by='TotalCoinsMined',ascending=False)
_,ax = plt.subplots(figsize=(20,5))
# plt.bar(x=data_trading_mined['CoinName'], height=data_trading_mined['TotalCoinsMined'])
sns.barplot(x='CoinName', y='TotalCoinsMined', data = data_plot.iloc[:50])
plt.xticks(rotation=90)
data_plot
```

```python
data_trading_mined.drop(['Unnamed: 0','CoinName'], axis=1, inplace=True)
data_trading_mined
```

```python
print(data_trading_mined[['Algorithm','ProofType']].nunique())
print('----------------------------------')
print(data_trading_mined.dtypes)
```

### TotalCoinSupply is in object dtype, convert to float

```python
data_trading_mined['TotalCoinSupply']= data_trading_mined['TotalCoinSupply'].astype('float')
data_trading_mined.dtypes
```

```python
data_trading_mined.head()
```

```python
X = pd.get_dummies(data_trading_mined, drop_first=True)
X
```

```python
X.shape
```

#### Initial data frame has 4 features. After one hot encoding, 71+25-2 = 94 features in Algorithm and Proof type, + 2 features in mined coins and total supply make it 96 features as shown in X.shape

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_scaled = StandardScaler().fit_transform(X)
```

### Scaled Data

```python
X_scaled
```

```python
from sklearn.decomposition import PCA
```

### Preserve 90% explained variance

```python
pca = PCA(n_components=0.9)
pca_data = pca.fit_transform(X_scaled)
```

```python
pd.DataFrame(data=pca_data)
```

### After PCA the number of dimensions reduced from 96 to 74 while preserving 90% expained variance

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

```python
from sklearn.manifold import TSNE
```

```python
# tsne=TSNE(perplexity = 30)
tsne=TSNE(perplexity = 15)
```

```python
tsne_features = tsne.fit_transform(pca_data)
tsne_features
```

```python
plt.scatter(tsne_features[:,0], tsne_features[:,1])
```

### Strong clustering no found in the data from t-SNE

```python
from sklearn.cluster import KMeans
```

```python

clusters = range(1,50)

inertia = []
for i in clusters:
    km = KMeans(n_clusters=i)
    km.fit(pca_data)
    inertia.append(km.inertia_)
```

```python
inertia
```

```python
plt.plot(clusters,inertia)
plt.ylabel('inertia')
plt.xlabel('no of clusters')
```

### Data does not show an elbow trend indicating there is no clustering
